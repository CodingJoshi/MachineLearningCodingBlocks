## Data visualisation Challenge
Given Movies Dataset,your taskie is to make a Visualisation that 
- plots the length of Movie tilte Name on x-Axis
- Frequency Counts on Y-axis i.e Numbers of Movies having 'x' characters in their title


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Read a CSV file
#data frame
df=pd.read_csv("MovieData.csv")
print(type(df))
```

```python
df.head() #be default it gives only 5 rows
# you can also specify
how many rows you want to show
df.head(n=10)
```

```python
df.columns
```

```python
titles=df.get('movie_title')
print(type(titles))
print(titles)

titles=list(titles)
print(type(titles))

print(titles[:5])
```

```python
freq_titles={}

for title in titles:
    length=len(title)
    
    if freq_titles.get(length) is None:
        freq_titles[length]=1
    else:
        freq_titles[length]+=1
```

```python
print(freq_titles)
```

```python
X=np.array(list(freq_titles.keys()))
Y=np.array(list(freq_titles.values()))
```

```python
# print(X,Y)

plt.scatter(X,Y)
plt.xlabel("Length of Movies Titles");
plt.ylabel("Freq Count")
plt.title("Movie Data Visualiation Problem")
plt.show()
```
