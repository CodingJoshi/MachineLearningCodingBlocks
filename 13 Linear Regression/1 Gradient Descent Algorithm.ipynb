# Gradient Desecent Algorithm

let us assume Loss/Error function is J(x)=(x-5)^2

**Goal** is to find X for which error is minimum
and also visualise gradient desent algorithm using plots



```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=np.arange(-25,36)
Y=(X-5)**2  # this is error function
print(X);
print();
print(Y);
```

```python
plt.style.use("seaborn")
plt.plot(X,Y)
plt.xlabel("X")
plt.ylabel("F(x)=(X-5)^2")

plt.show();
```

```python
def Gradient(x):
    return 2*(x-5);

def GradientDescent(X,Y):
    # assume intial value of x 
    x=30.0
    #learning rate=0.1
    lr=0.1;
    cnt=0;
    stepsize=lr*Gradient(x);
    while stepsize>0.001:
        g=Gradient(x);
        stepsize=lr*g;
#         print(stepsize)
        x=x-stepsize
        cnt+=1;
    print(x);
```

```python
GradientDescent(X,Y)
```
