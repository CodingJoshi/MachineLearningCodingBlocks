```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
X=pd.read_csv("./xdata.csv").values;
Y=pd.read_csv("./ydata.csv").values;
print(X.shape,Y.shape)
```

```python
X=X[:,1:]
Y=Y[:,1:].reshape((-1,))
print(X.shape,Y.shape)
# print(X);
# print(Y);
```

```python
#visualization
plt.style.use('seaborn')
plt.scatter(X[:,0],X[:,1],c=Y);
plt.show()
```

```python
def distance(x1,x2):
    return np.sqrt(sum((x2-x1)**2));
```

```python
def KNN(X,query_point,k=5):
    vals=[];
    for i in range(X.shape[0]):
        vals.append((distance(X[i],query_point),Y[i]))
        
    vals=sorted(vals);
    vals=vals[:k];
#     print(vals);
    vals=np.array(vals);
    freq=np.unique(vals[:,1],return_counts=True);
    indx=np.argmax(freq[1]);
    pred=freq[0][indx];
    return pred;
    
```

```python
# now plot query point and try to predict it's class using kNN
#visualization
query_point=[0,1];
plt.scatter(X[:,0],X[:,1],c=Y);
plt.scatter(query_point[0],query_point[1],color='red');
plt.show()
```

```python
grp=KNN(X,query_point);
print(grp)
```
